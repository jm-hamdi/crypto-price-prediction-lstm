```python
# Charger les données (ex: CSV OHLCV)
def load_data_csv(path):
    df = pd.read_csv(path, parse_dates=['timestamp'])
    df = df.sort_values('timestamp').set_index('timestamp')
    # garder close et volume
    df = df[['close', 'volume']].ffill()
    return df
```
